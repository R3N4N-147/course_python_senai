```python
# **5.** O programa deve: responder se o aluno foi aprovado. 
# Receber os dados de um aluno que será do senai vila alpina ou do mooca. 
# Se ele for do senai vila alpina, no curso de python ele deve ter pelo menos 60 de nota e 75% de frequência. 
# Se ele for do senai mooca, também do curso de python, ele deve ter a mesma quantidade de presença porém 
# com uma nota de 50 ele já consegue ser aprovado.
# *Dados: O curso possui 80 aulas. Input: número de faltas*

un = input("Curso Python. Informe a unidade SENAI informando a LETRA + enter. (V: Vila Alpina. M: Mooca)")
grade = float(input("Informe a nota final do aluno."))
freq = int(input("Informe a quantidade de aulas em que o aluno faltou. (Use números inteiros)"))
aulas = 80

if un == "V":
    if grade >= 60 and freq >= aulas*0.75:
        print("Aprovado")
if un == "M":
    if grade >= 50 and freq >= aulas*0.75:
        print("Aprovado")
else:
    print("Reprovado")
```

```python
# 5. (melhorado)
un = input("Curso Python. Senai Unidade V ou M? (V: Vila Alpina. M: Mooca)")
grade = float(input("Informe a nota final do aluno."))
freq = int(input("Informe a quantidade de aulas em que o aluno faltou. (Use números inteiros)"))
aulas = 80

if freq < aulas*0.75: ## Requisito de eliminação para ambos
    if un == "M" and grade >= 50: ## requisitos de eliminação individuais
        print("Aprovado")
    if un == "V" and grade >= 60: ## requisitos de eliminação individuais
        print("Aprovado")
    else:
        print("Reprovado. Nota insuficiente.")
else:
    print("Reprovado por baixa frequencia.")
```

```python
# 7. Imagine que você está numa escola. Diga se o aluno está aprovado se:
# a) A nota dele for maior ou igual a 50.
# a.a) Faça também para notas superiores a 30, ser possível fazer uma recuperação.
# b) A nota dele for maior ou igual a 50 e a sua frequência superior a 75% da quantidade de horas de curso
# c) A nota dele for superior a 70 (caso seja inferior a 30 assuma que ele está reprovado, entre 30 e 50 está de recuperação
# e acima de 50 até 70 está de exame). 
# E também, se for da escola HRC ele precisa ter 50% de frequência. 
# Se for de outra escola, 40%.
# Nome ,  Nota  ,  Quantidade de horas do curso (h)   ,   Faltas (h)

escola7 = input("Informe o nome da escola (HRC ou outra)")
nota7 = float(input("Informe a nota final do aluno"))
hora7 = int(input("Informe o quantidade de horas do curso"))
falta7 = int(input("Informe o total de faltas em horas."))

print("\nDiga se o aluno está aprovado se:\n")
print("Ex.7.a: A nota dele for maior ou igual a 50.")
print("Ex.7.a.a: Faça também para notas superiores a 30, ser possível fazer uma recuperação.")

if nota7 > 30:
    if nota7 >= 50:
        print("Aprovado")
    if nota7 < 50 and nota7 > 30:
        print("Recuperacao")
if nota7 <= 30:
    print("Reprovado")

print("\nEx.7.b: A nota dele for maior ou igual a 50 e a sua frequência superior a 75% da quantidade de horas de curso")
if nota7 >= 50 and (hora7-falta7) > hora7*0.75:
    print("Aprovado.")
else:
    print("Reprovado.")

print("\nEx.7.c: A nota dele for superior a 70 \n(caso seja inferior a 30 assuma que ele está reprovado,")
print("\nentre 30 e 50 está de recuperação e \nacima de 50 até 70 está de exame).")
print("\nE também, se for da escola HRC ele precisa ter 50% de freq. Se for outra esccola, 40%.")

if escola7 = "HRC":
    if (hora7-falta7) > hora7*0.50:
        if nota7 > 70:
            print("Aprovado")
        if nota7 <= 70 and nota7 > 50:
            print("Exame")
        if nota7 <= 50 and nota7 >= 30:
            print("Recuperacao")
        if nota7 < 30:
            print("Reprovado")
    else:
        print("Reprovado por frequencia. Menor 50%")
else: ## outras
    if (hora7-falta7) > hora7*0.40:
        if nota7 > 70:
            print("Aprovado")
        if nota7 <= 70 and nota7 > 50:
            print("Exame")
        if nota7 <= 50 and nota7 >= 30:
            print("Recuperacao")
        if nota7 < 30:
            print("Reprovado")
    else:
        print("Reprovado por frequencia. Menor 40%")
```
