```python
## 4. Numa situação real do dia a dia de Adriana, ela precisa fazer o orçamento de 5 itens para seus clientes, 
# e mostrar uma mensagem com o total. Faça um programa que resolva esta situação simples de soma pra ela, 
# imprimindo a mensagem, ao final: <O total do orçamento foi R${total}>

item1 = float(input("Adicione item 1 (por ex.: 20.00)"))
item2 = float(input("Adicione item 2 (por ex.: 20.00)"))
item3 = float(input("Adicione item 3 (por ex.: 20.00)"))
item4 = float(input("Adicione item 4 (por ex.: 20.00)"))
item5 = float(input("Adicione item 5 (por ex.: 20.00)"))
sum = item1 + item2 + item3 + item4 + item5
print("item 1: R$ {}\nitem 2: R$ {}\nitem 3: R$ {}\nitem 4: R$ {}\nitem 5: R$ {}\n".format(item1,item2,item3,item4,item5))
print("O total do orcamento foi R$ {:.2f}".format(sum))
```

```python
## 4. (Melhorado) Numa situação real do dia a dia de Adriana, ela precisa fazer o orçamento de 5 itens para seus clientes, 
# e mostrar uma mensagem com o total. Faça um programa que resolva esta situação simples de soma pra ela, 
# imprimindo a mensagem, ao final: <O total do orçamento foi R${total}>
lista_itens = []

for i in range(5):
    valor = float(input(f"Adicione o valor do item {i+1} (por ex.: 20.00)"))
    lista_itens.append(valor)

print(f"O total do orçamento foi R${sum(lista_itens)}")
```

```python
## 7. Crie um script em python que pergunte o nome, a idade, o salário e as despesas mensais de um indivíduo 
# utilizando a função input(). Crie, então, uma resposta que diga:
# "O <nome> recebe <salario> tendo <idade> anos e tem <restante> disponível ao fim do mês"

nome = input("Qual seu nome? ")
idade = input("Qual sua idade? ")
salario = float(input("Qual seu salário? (com centavos, separados por ponto) "))
despesas = float(input("Quais suas despesas mensais? (com centavos, separados por ponto) "))
restante = salario - despesas
print(f"O {nome} recebe R$ {salario} tendo {idade} anos e tem R$ {restante} disponível ao fim do mês.")
```
