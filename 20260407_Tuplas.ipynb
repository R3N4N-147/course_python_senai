ex.2. Unpacking de Coordenadas: Imagine que uma máquina de GPS te entregou uma tupla com a latitude e a longitude: local = (-23.55, -46.63). Use o desempacotamento para colocar esses valores em duas variáveis chamadas lat e log e imprima cada uma.

```python
local = (-23.55, -46.63)

# unpacking
lat, log = local
print(f"Latitude: {lat}")
print(f"Longitude: {log}")
```

ex.3. Buscando na Tupla: Crie uma tupla com as vogais. Peça para o usuário digitar uma letra e use o método .count() para dizer quantas vezes aquela letra aparece e o método .index() para dizer em qual posição ela está (caso exista).


```python
vogais = ('a', 'e', 'i', 'o', 'u') # tupla
letra = input("Digite uma vogal: ").lower()

if letra in vogais:
    print(f"A letra aparece {vogais.count(letra)} vez(es).")
    print(f"Ela está na posição: {vogais.index(letra)}")
else:
    print("Isso não é uma vogal na nossa tupla.")
```

 Distribuição de Dados: Você recebeu os dados de um cliente em uma tupla: cliente = ("Marcelo", 35, "São Paulo"). Realize o unpacking desses dados e imprima uma frase formatada.

```python
cliente = ("Marcelo", 35, "São Paulo")

nome, idade, cidade = cliente

print(f"Cliente: {nome}, {idade} anos, cidade de origem: {cidade}.")
```

Ex.5  O Valor Ignorado: Dada a tupla cores = ("Vermelho", "Verde", "Azul"), realize o unpacking, mas use o caractere _ (underscore) para a cor do meio. O _ é uma convenção no Python para dizer: "Eu sei que esse valor existe, mas vou ignorá-lo".

```python
cores = ("Vermelho", "Verde", "Azul")

primeira, _, ultima = cores

print(f"A primeira cor é {primeira} e a última é {ultima}.")
```
