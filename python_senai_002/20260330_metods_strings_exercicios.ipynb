```python
# 1. Crie um programa em python que verifique se o email do usuário possui um “@”, 
# e caso haja, escreva “Cadastro efetuado com sucesso.” Caso contrário, escreva “Endereço de email inválido”.

s1 = input("Insira seu endereço de email completo.")
if s1.find("@") > 0:
    print("Cadastro efetuado com sucesso.")
if s1.find("@") < 0:
    print("Endereço de email inválido.")
```

```python
# **2.** Usando como base o programa anterior, se houver um “@”, verifique se há um “.” após. 
#     Em caso de erro, indique o motivo da falha, e em caso de acerto, escreva “Cadastro efetuado com sucesso”
# Desafio: Indique também o domínio do email.

# Essas frases me matam ... acho que ele quis dizer pra:
# 1- verificar se tem '@' (ex. 1)
# 2- verificar se depois do arroba existe em qualquer posicao algum '.'
# Como essa etapa é sobre strings, vou usar conceitos de metodos de strings...
# Anteriormente pesquisei como mostrar uma string a partir de uma posicao retornada

string = input("Insira seu endereço de email completo.")
p = string.find("@") ## busca posicao
if p > 0:
    if string.find(".") > 0:
        print("Cadastro efetuado com sucesso.\nDominio: {}".format(string[p:]))
    else:
        print("Pontuação '.' não encontrada.")
else:
    print("Endereço de email inválido.")
```

```python
## **2.** (acho que o melhor no fim é usar o 'in')

string = input("Insira seu endereço de email completo.")
p = string.find("@") # acha posicao do @
if "@" in string:
    if "." in string[p:]: # fiz [p:] pq meu email tem ponto '.' antes do '@', e eu quero '.' após o '@'
        print("Cadastro efetuado com sucesso.\nDominio: {}".format(string[p+1:]))
    else:
        print("Pontuação '.' não encontrada.")
else:
    print("Endereço de email inválido.")
```

```python
# 3. Faça um programa que peça o cadastro de um CPF. O programa deve remover os espaços, pontos e hifens. 
#     Se o CPF tiver 11 números após a remoção de pontos e hifens, o CPF deve ser aprovado.

cpf = input("Informe CPF:")
cpf = cpf.replace('.','').replace('-','').replace(' ','')
# print(type(cpf))
if len(cpf) == 11 and cpf.isnumeric():
    print("Aprovado. CPF: {}".format(cpf))
else:
    print("Error")
```
