````python
tarefas = {1: "Lavar louça", 2: "Estudar Python", 3: "Treinar"}
concluida = int(input("Qual tarefa você concluiu? (1-3): "))

# O .pop() remove e nos "entrega" o valor removido na variável 'item'
if concluida in tarefas:
    item = tarefas.pop(concluida)
    print(f"A tarefa '{item}' foi concluída e removida com sucesso!")
else:
    print("Número de tarefa inválido.")
````

4.2 Baseado no programa anterior, melhore-o. Agora, faça com que sua agenda, além de números, guarde também e-mails, endereços, telefone, parentesco.

Opções no programa:

1 - Adicionar novo contato 2 - Consultar contato 3 - Ver lista de contatos 4 - Remover um contato 5 - Fechar

Faça também com que seu programa tenha a opção de mostrar agenda completa, realizando um print no formato:

     Leonardo:

Telefone: 111111

Endereço: rua flores

E-mail: Leozito@bol

relação: Amigo

````python
agenda = {}

while True:
  op = input("1-Adicionar 2-Consultar 3-Ver Lista 4-Remover 5-Sair\nEscolha: ")
  match op:
    case "1":
      nome = input("Nome: ").title()
      # Criamos um dicionário interno para guardar os detalhes
      detalhes = {
          "Telefone": input("Telefone: "),
          "Endereço": input("Endereço: "),
          "E-mail": input("E-mail: "),
          "Relação": input("Relação: ")
          }
      agenda[nome] = detalhes
      print(f"Contato {nome} adicionado!")
    case "2":
      nome = input("Nome para consultar: ").title()
      if nome in agenda:
        print(f"\n{nome}:")
        for campo, valor in agenda[nome].items():
          print(f"{campo}: {valor}")
        else:
          print("Contato não encontrado.")
    case "3":
      if not agenda:
        print("Agenda vazia.")
      else:
        for nome, info in agenda.items():
          print(f"\n{nome}:")
          print(f"  Telefone: {info['Telefone']}")
          print(f"  Endereço: {info['Endereço']}")
          print(f"  E-mail: {info['E-mail']}")
          print(f"  Relação: {info['Relação']}")
    case "4":
      nome = input("Nome para remover: ").title()
      if nome in agenda:
        del agenda[nome]
        print(f"Contato {nome} removido.")
      else:
        print("Contato não encontrado.")
    case "5":
      break
    case _:
      print("Opção inválida.")
````
