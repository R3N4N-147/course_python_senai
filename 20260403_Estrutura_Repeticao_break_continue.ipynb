```python
for numero in range(10): ## Comando que inicia em 0 e vai até 10
    if numero == 5:
        break  # Interrupção. Sai do loop quando numero é 5
    print(numero)
# Saída: 0, 1, 2, 3, 4
```

```python
for numero in range(5):
    if numero == 2:
        continue  # Pula a impressão quando numero é 2
    print(numero)
# Saída: 0, 1, 3, 4 (2 foi pulado)
```

```python
# Crie um while True (loop infinito) que peça o nome de uma fruta. Se o usuário digitar "sair", use o break para encerrar o programa.


while True:
    fruta = input("Digite uma fruta (ou 'sair' para parar): ").lower()
    if fruta == "sair":
        break
    print(f"Adicionando {fruta} à cesta...")

```

```python
# Pulando Números com Continue: Use um for de 1 a 10 e imprima todos os números, exceto os números 3 e 7, usando o comando continue.

for i in range(1, 11):
    if i == 3 or i == 7:
        continue # Pula a execução e volta para o próximo número
    print(i)
```
