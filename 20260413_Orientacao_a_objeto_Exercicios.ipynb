(Transição): O Objeto que "Fala"
Crie uma classe Lampada com o atributo ligada = False. Crie os métodos acender() e apagar() que alteram esse valor e imprimem o status atual.
Objetivo: Praticar a alteração de atributos internos através de métodos

````python
# # Definindo o molde (Classe)
# class Contato:
#     def __init__(self, nome, telefone, email):
#         # Atributos (o que o contato TEM)
#         self.nome = nome
#         self.telefone = telefone
#         self.email = email

#     # Método (o que o contato FAZ)
#     def enviar_email(self, mensagem):
#         print(f"Enviando para {self.email}: {mensagem}")

class Lampada:
    # 1. O nascimento (sempre começa desligada)
    def __init__(self, estado = False): # criar o objeto
        self.ligada = estado # preciso declarar mesmo com 1 atributo, e mesmo sendo fixo.

    # 2. Método para ACENDER
    def acender(self):
        self.ligada = True
        print("A lâmpada foi ACESA! 💡")

    # 3. Método para APAGAR
    def apagar(self):
        self.ligada = False
        print("A lâmpada foi APAGADA! 🌑")

minha_lampada = Lampada() # Salvando o objeto (a lampada) em uma var ...
# Interaja com o objeto atraves dos métodos
# Muito parecido com os métodos de string, métodos de listas, ou outro ... variavel_armazenada.metodo()
print(f"A lâmpada está ligada? {minha_lampada.ligada}\n") # False

minha_lampada.acender() # Chama a ação
print(f"E agora 1? {minha_lampada.ligada}\n") # True

minha_lampada.apagar() # Chama a ação
print(f"E agora 2? {minha_lampada.ligada}") # False
````

````python
# Ex. 10
# extra
# sobre "navegar" dentro de dicionarios

class Carteira: # classe
  def __init__(self, dono): # Parametros
    # Atributos
    self.dono = dono
    self.cartoes = [] # lista de dicionarios

  # Métodos (ações)
  def adicionar_cartao(self, nome, banco):
    novo_cartao = {"nome": nome, "banco" : banco}
    self.cartoes.append(novo_cartao)

  def buscar_banco(self, banco_alvo):
    achou = False
    for cartao in self.cartoes: # navega na lista armazenada em self.cartoes
      # Olha cada chave "banco" e compara o valor armazenado com o valor de buscar_banco
      if cartao["banco"] == banco_alvo:
          print(f"Nome: {cartao['nome']} : Banco: {cartao['banco']}")
          achou = True
    if not achou:
       print(f"Nenhum cartão do banco {banco_alvo} encontrado.")

  def listar_bancos(self):
    print(f"Lista de bancos cadastrados de {self.dono}:")
    if not self.cartoes:
      print("Nenhum cartão cadastrado.")
    else:
      for k in self.cartoes:
        # k é o dicionario atual: {"nome": "...", "banco": "..."}
        print(f"Nome: {k["nome"]}, Banco: {k["banco"]}")


# TESTES
wallet_1 = Carteira("Renan")
wallet_1.adicionar_cartao("Visa Gold", "Nubank")
wallet_1.adicionar_cartao("Mastercard", "Inter")
wallet_1.adicionar_cartao("Ultravioleta", "Nubank")

wallet_1.buscar_banco("Nubank")
print()
wallet_1.listar_bancos()
````
