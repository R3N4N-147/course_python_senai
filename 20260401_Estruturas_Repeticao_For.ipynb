```python
# Podemos ter um 'for' percorrendo uma lista iteravel

listaexemplo = [1, 2, 'a', 'b', 'c', 'd']
# Aqui estou atribuindo ao item, a cada iteração, o valor atual da lista começando pelo primeiro valor até o último.
for item in listaexemplo:
		# Este método é interessante quando temos somente uma lista, pois é mais ágil do que fazer: for item in range(len(listaexemplo)) print(listaexemplo[item])
		print(item)
```

```python
cadastros = []

for i in range(4):
    porsche = input('Digite o nome a ser cadastrado\n')
    cadastros.append(porsche)
print('a lista de nomes é {}, {}, {}, {}'.format(cadastros[0],cadastros[1],cadastros[2],cadastros[3]))
```

```python
listaexemplo = ['a', 'k', 'c', 'd']

# Aqui, conseguimos salvar tanto o índice, quanto o item da lista
for i, j in enumerate(listaexemplo):
    print(i)
    print(j)
```

```python
# Criando listas com o “for”
```

```python
# list comprehension
listanova = [i for i in range(5)] # Aqui criamos as iterações do for somente dentro da lista
print(listanova)
```

```python
# Aqui conseguimos criar a lista com um input diretamente do for
listanova = [input('Digite algo: ') for i in range(4)]
print(listanova)
```

```python
lista = [i for i in range(20) if i%2==0]
print(lista)
# E, caso eu tivesse uma função que me entregasse o valor desejado para cada índice da lista, poderia substituir o i pelo nome da função
```
