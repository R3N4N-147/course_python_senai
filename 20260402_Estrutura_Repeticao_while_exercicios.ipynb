```python
# Utilizando a biblioteca random > randint, crie um loop while
# que peça ao usuário que adivinhe um número entre 1 e 999, caso ele
# erre o número diga se o resultado é menor ou maior.

from random import randint
r = randint(1,999)
guess = -1

print("Digite 0 para sair.")
print("Adivinhe o número entre 1 e 999:")

while guess != r:
  try:
    guess = int(input(" "))

    if guess == 0:
      print("Fim de jogo.")
      break
    else:
      if guess > r:
        print(f"É menor do que {guess}")
      else:
        if guess < r:
          print(f"É maior do que {guess}")
        else:
          if guess == r:
            print(f"Grande vencedor! Número {guess}. Acertou!")
  except ValueError:
    print("Invalido.")
```

```python
# resolucao
from random import randint
r = randint(1,1000)
guess = -1 # Pode ser qualquer numero (Lembrar de testar com true)
count = 0

print("Digite 0 para sair.")
print("Adivinhe o número entre 1 e 1000:")


while guess != r:
  try:
    guess = int(input(" "))
    count += 1

    if guess == 0:
      print("Fim de jogo.")
      break
    if count == 12: # 12 chances
      print("Fim das chances.")
      break
    else:
      if guess > r:
        print(f"É menor do que {guess}")
      else:
        if guess < r:
          print(f"É maior do que {guess}")
        else:
          if guess == r:
            print(f"Grande vencedor! Número {guess}. Acertou!")
  except ValueError: # esta dentro do while
    print("Invalido.")
```

```python
# Desafio 1
# Refatorado
f = []
p = []
while True:
  op = int(input(" 1: Consulta\n 2: Adicionar\n 3: Remover itens\n 4: Calcular preço total\n 5: Sair\n"))
  match op:
    case 1:
      if len(f) == 0:
        print("Lista vazia.")
      else:
        for i in range(len(f)):
          print(f"Lista de compras: {f[i]} R${p[i]:.2f}")
    case 2:
      nome_f = input("Fruta: ").title()
      valor_p = float(input("Preço: "))
      f.append(nome_f)
      p.append(valor_p)
    case 3:
      item = input("Digite a fruta a ser removida: ").title()
      if item in f:
        ind = f.index(item)
        f.pop(ind)
        p.pop(ind)
      else:
        print("Fruta não encontrada.")
    case 4:
      print(f"Preço total: {sum(p):.2f}")
    case 5:
      break
    case _:
      print("Opção inválida.")
```

```python
# Desafio 2
# Pergunte também o salário e a quantidade vendida, dizendo assim,
# quanto % do salário está sendo gasto em mercado
f = []
p = []
q = [] # Lista quantidade vendida

salario = float(input("Qual o seu salário liquido mensal? R$ ")) # pode dar problema caso digite outra coisa

while True:
  op = int(input(" 1: Consulta\n 2: Adicionar\n 3: Remover itens\n 4: Calcular preço total e % sobre salário\n 5: Sair\n"))

  match op:
    case 1:
      if not f:
        print("Lista vazia.")
      else:
        print("Lista de compras:\n")
        for i in range(len(f)):
          print(f"{q[i]} x {f[i]} à R${p[i]:.2f} (subtotal: R$ {p[i]*q[i]:.2f})")
    case 2:
      nome_f = input("Fruta: ").title()
      valor_p = float(input("Preço: "))
      quantidade_q = int(input("Quantidade: "))
      f.append(nome_f)
      p.append(valor_p)
      q.append(quantidade_q)
    case 3:
      item = input("Digite o nome da fruta a ser removida: ").title() # "item"
      if item in f:
        ind = f.index(item) # Todas as listas possuem o mesmo indice para o mesmo item
        f.pop(ind) # lista com frutas
        p.pop(ind) # lista com preços
        q.pop(ind) # lista com quantidades
        print(f"Item {item} removido.")
      else:
        print("Fruta não encontrada.")
    case 4:
      total = 0 # Zerar a lista antes ...
      for i in range(len(f)):
        total += p[i]*q[i] # preço x quantidade
      print(f"Preço total: {total:.2f}")
      porcentagem = (total/salario)*100
      print(f"Na atual compra mercado reprenta {porcentagem:.2f}% do seu salário.")
    case 5:
      break
    case _:
      print("Opção inválida.")
```
