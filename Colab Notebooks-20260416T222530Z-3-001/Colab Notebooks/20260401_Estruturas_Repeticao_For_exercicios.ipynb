```python
# 8. Faça um programa que, para uma lista de mercado de 10 produtos
# pré-defnidos por você, gere preços aleatórios (incluindo centavos).
# A partir deste, crie um sistema de mercado, onde a pessoa via um menu
# pode adicionar novos produtos, consultar preços de produtos já existentes,
# verificar o catálogo de produtos e excluir produtos existentes.

mercado = ["Pão", "Queijo", "Presunto", "Leite", "Café", \
           "Açúcar", "Arroz", "Feijão", "Papel higiênico", "Detergente"]
precos = [4.50, 12.90, 8.45, 5.20, 18.30, 22.15, 9.80, 4.75, 11.20, 25.60]

for i in range(len(mercado)):
  print(f"{i+1}", mercado[i])

print("\nMenu:")
print("1: cadastro de produto.")
print("2: Descadastro de produto pelo nome.")
print("3: Descadastro de produto pelo número.")
print("4: Consultar preco.")
print("99: Para sair.\n")

SN = int(input())
match SN:
  case 1:
    c1 = input("Nome do novo produto: ")
    mercado.append(c1)
    precos.append(int(input("Preco do novo produto: ")))
    #print(mercado.index(c1))
  case 2:
    c2 = input("Descadastro.\nNome do produto: ")
    if c2 in mercado:
      ind = mercado.index(c2)
      mercado.pop(ind)
      precos.pop(ind)
    else:
      print("Nome encontrado.")
  case 3:
    c3 = int(input("Descadastro.\nNúm. do produto: "))
    if c3 >= 0 or c3 <= len(mercado):
      mercado.pop(c3-1)
      precos.pop(c3-1)
    else:
      print("Número não encontrado.")
  case 99:
    print('libra') #
  case _:
    print('nenhum')

for i in range(len(mercado)):
  print(f"{i+1}", mercado[i])
```

```python
# Desafio
# **1.** A partir do exercício 6, pesquise as faixas de imposto do INSS e utilize a
#  porcentagem adequada para cada faixa salarial.

# O emunciado é estranho ...

sal = float(input("Informe o salário. Use ponto para separar os centavos."))
if sal <= 2428.80:
    print(f"Salario R${sal:.2f}, aliquota: Isento.")
if sal >= 2428.81 and sal <= 2826.65:
    print(f"Salario R${sal:.2f}, aliquota 7.5%: {sal*0.075:.2f}.")
if sal >= 2826.26 and sal <= 3751.05:
    print(f"Salario R${sal:.2f}, aliquota 15%: {sal*0.15:.2f}.")
if sal >= 3751.06 and sal <= 4664.68:
    print(f"Salario R${sal:.2f}, aliquota 22.5%: {sal*0.225:.2f}.")
if sal > 4664.68:
    print(f"Salario R${sal:.2f}, aliquota 27.5%: {sal*0.275:.2f}.")

```
