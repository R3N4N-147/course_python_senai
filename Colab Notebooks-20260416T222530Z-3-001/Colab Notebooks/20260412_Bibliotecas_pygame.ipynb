```python
# As importações de bibliotecas que utilizaremos serão as:

import pygame #importa a biblioteca toda do pygame
from pygame.locals import * #importa todo o conteúdo da "pasta" locals
from sys import exit # importa a funcionalidade de fechar do sistema (botão fechar)

# Inicializaremos o módulo do pygame, criando então a nossa tela e dando um nome a ela:

pygame.init()
larg,alt = 1920,1080 # definindo tamanho da tela
tela = pygame.display.set_mode((larg,alt)) #Largura e altura em pixels da sua tela, sendo números inteiros
pygame.display.set_caption('Lobo Solitário (Lone Wolf by Joe Dever)')

#Todo jogo se passa dentro de um loop, que é representado pela frequência de atualização
while True:
    for event in pygame.event.get(): #pega os eventos da fila
    #https://www.pygame.org/docs/ref/event.html
        if event.type == QUIT:
            pygame.quit() #sair do pygame
            exit() #fechar a janela

    pygame.draw.rect(tela, (10,110,10),(200,300,40,50)) # (10,10,10 tupla rgb),(200,300,#LxH onde vai ser desenhado 40,50 LxH do rect sendo referenciado no canto superior esquerdo
    # tela: Superfície aonde o objeto será desenhado
    # (10,110,10) tupla com as cores do objeto no espectro rgb
    pygame.draw.circle(tela,(10,110,10),(300,300),15) # (300,300)pos,15 raio no centro)
    # pygame.draw.circle(tela,(10,110,10),(posx1,posy1),raio)
    # posição em x e y e o raio do círculo
    pygame.draw.line(tela,(10,110,110),(100,100),(150,150),3) # (100,100)pos ini,(150,150) pos fim,3 espessura
    pygame.display.update() # realiza um refresh na tela
```
