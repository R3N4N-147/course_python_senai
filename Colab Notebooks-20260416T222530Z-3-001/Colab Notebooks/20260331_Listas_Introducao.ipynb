```python
# Estas duas listas mostram diferentes produtos e suas vendas, respectivamente.
produtos = ['cerveja','arroz', 'mortadela','miojo','old eight']
vendas = [100,35,330,20,100]

produto = 'arroz'
i = produtos.index(produto)
print('O produto {} teve um total de {} vendas.'.format(produtos[i], vendas[i]))
print(f'O produto {produto[i]} teve um total de {vendas[i]} vendas.')
```
