```python
class Cachorro:
  def __init__(self,nome,raca = "vira lata"):
    self.nome = nome
    self.raca = raca
    self.idade = 0
  def informacoes(self):
    return f'''Nome: {self.nome}
Raca: {self.raca}
Idade: {self.idade}'''

cachorros = []
cachorros.append(Cachorro("Zeus","Bullterrier"))
cachorros.append(Cachorro("Thor"))
cachorros.append(Cachorro("Nina","Spitz"))
nome = input("Informe nome do cachorro: ")
for cachorro in cachorros:
  if cachorro.nome == nome:
    print(cachorro.informacoes())
```
