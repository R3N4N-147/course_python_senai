Ex.1. (Transição): O Molde Simples.
Crie uma classe chamada Pessoa que receba apenas nome e idade. Instancie dois objetos e imprima: "Olá, meu nome é [nome]".
Objetivo: Entender a sintaxe básica e o uso do self.

```python
# Definindo o molde (Classe)
class Pessoa:
    def __init__(self, nome, idade):
        # Atributos (o que o contato TEM)
        self.nome = nome
        self.idade = idade

# Instancie (criar um objeto daquela classe) dois objetos e imprima uma mensagem
pessoa1 = Pessoa("Hänsel", 9)
pessoa2 = Pessoa("Gretel", 7)

print(pessoa1.nome, pessoa1.idade)
print(pessoa2.nome, pessoa2.idade)

print(pessoa1)

print(f"Olá, meu nome é {pessoa1.nome}")
print(f"Olá, meu nome é {pessoa2.nome}, tenho {pessoa2.idade} anos de idade.")
```

(Proposta do professor): Criando uma classe chamada cachorro.

Crie 5 objetos (cachorros) diferentes ( Em outras palavras: Instanciar 5 objetos).

```python
class Cachorro: # classe Cachorro
    def __init__(self, raca, cor, nome, idade, peso): # ( "aqui dentro" são os parametros )
        # Atributos
        self.raca = raca
        self.cor = cor
        self.nome = nome
        self.idade = idade
        self.peso = peso
    # poderia criar um método aqui
    # Imagine um código de método aqui, mais nos proximos captulos :)

# instanciando 5 objetos (criando)
cachorro1 = Cachorro("rottweiler","castanho","nenem",0,16.5)
Cachorro2 = Cachorro("pug","misto preto e branco","lola",0,6)
Cachorro3 = Cachorro("pastor alemao","red and black","Tulipa",1,25)
Cachorro4 = Cachorro("beagle","red tricolor","bolacha",1,10)
Cachorro5 = Cachorro("bulldog","fawn and white","Handsome Dan",1,18)

# testes de impressão
# usando o método de string print().
# print(self.atributo)
print(cachorro1.raca,cachorro1.cor,cachorro1.nome,cachorro1.idade,cachorro1.peso)
# usando o método print(f-strings) (Mais eficiente...)
print(f"{cachorro1.raca} {cachorro1.cor} {cachorro1.nome} {cachorro1.idade} {cachorro1.peso}")
print(f"Raça: {cachorro1.raca}, Cor: {cachorro1.cor}, Nome: {cachorro1.nome}, Idade: {cachorro1.idade}, Peso: {cachorro1.peso} kg.")

# usando o laço for para evitar repetição (Ainda mais eficiente...)
# É preciso ter os objetos criados em uma lista
# Eu poderia ter criado os objetos direto dentro da lista
# No meu caso vou chamar os objetos já criados.
lista_matilha = [cachorro1,Cachorro2,Cachorro3,Cachorro4,Cachorro5]
print("\n")
for c in lista_matilha:
    print(f"Nome: {c.nome} | Raça: {c.raca} | Cor: {c.cor} | Idade: {c.idade} ano(s) | Peso: {c.peso}kg")
print("\n")
for c in lista_matilha:
    print(f"Nome: {c.nome:<12} | Raça: {c.raca:<15} | Cor: {c.cor:<20} | Idade: {c.idade:<2} ano(s) | Peso: {c.peso:<2}kg")

print("\n")
print("Atenção ao detalhe ... vamos ver isso com mais clareza depois.")
print(cachorro1) # acessando o endereço de memoria onde o valor da var está salva.
```

```python
# exemplo de resposta:
# Se você tentar imprimir algo após criar classe, digamos no exemplo anterior, vamos acessar ao indesejado. Eis a saída:

class Cachorro:
    def __init__(self, raca, cor, nome, idade, peso):
        self.raca = raca
        self.cor = cor
        self.nome = nome
        self.idade = idade
        self.peso = peso

    # Este método diz ao Python como "converter" o objeto em texto
    def __str__(self):
        return f"Cão: {self.nome} | Raça: {self.raca} | Peso: {self.peso}kg"

# Instanciando
dog = Cachorro("Beagle", "Tricolor", "Bolacha", 1, 10)

# Agora você pode imprimir o objeto DIRETAMENTE!
print(dog)
```
