```python
# ATENÇÃO

# O que NÃO é Alfanumérico?
# Qualquer caractere que seja um símbolo, pontuação ou espaço. 
# Símbolos: @, #, $, %, &, *, +.
# Pontuação: ., ,, !, ?, -, _.
# Espaços: O próprio espaço em branco ( ) não é alfanumérico.
# Exemplos: renan@email.com (não é alfanumérico por causa do @ e do .).
# obs.: pesquise online ou veja a ASCII no prompt para relembrar o que é alphanumerico.
```

```python
print("Uma_frase".isnumeric())
print("Um dois".isnumeric())
print("123".isnumeric())
print("1 2 3".isnumeric())
print("1,2,3".isnumeric())
```

```python
print("Uma_frase".isalnum())
print("Um dois".isalnum())
print("Senai123".isalnum())
print("SenaiVILA".isalnum())
print("Senai VILA".isalnum())
print("1 2 3".isalnum())
print("1,2,3".isalnum())
```
