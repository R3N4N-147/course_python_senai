```python
try:
		b = int(input("Digite um número que será dividido por dois"))
		a = b/2
		print(a)
except:
		print('Erro.')
```

```python
try:
    a = 2/'s'
except TypeError:
		# Aqui, o erro é causado por um erro de digitação.
    print('O valor não é um número inteiro.')
except ZeroDivisionError:
		# Aqui, o erro é causado pela divisão por zero.
    print('O valor não pode ser dividido por zero.')
```

```python
## TESTE ALEATORIO
nome = "Renan da Fonseca"
print(f"nome[0:]: {nome[0:]}")
print(f"nome[:]: {nome[:]}")
print(f"nome[:len(nome)]: {nome[:len(nome)]}")
print(f"nome[1:len(nome)]: {nome[1:len(nome)]}")
## print(nome[-len(nome):0]) ## Nao printa nada
## print(nome.reverse()) ## Nao

print(f"\nnome[::-1]: {nome[::-1]}")
```
